## 3b. Missing values 

```python
# Checking for missing values
missing_values=df.isnull().sum()

# Calculate the total number of rows in the DataFrame
total_cells = df.shape[0]

# Calculate the percentage of missing values
percent_missing = (missing_values / total_cells) * 100

# Round the decimal to 2 decimal places
percent_missing = percent_missing.round(2)

# Convert the float type to string
percent_missing = percent_missing.astype(str)

# Add the '%' symbol to each value
percent_missing = percent_missing + '%'

# Combine the two Series into a new DataFrame
missing_value_table = pd.concat([missing_values, percent_missing], axis=1)

# Rename the columns of the DataFrame
missing_value_table.columns = ['Missing Values', 'Percent Missing']

# Sort the resulting DataFrame based on the 'Percent Missing' column in ascending order
missing_value_table.sort_values('Percent Missing', ascending=True, inplace=True)

# Print the resulting table
print(missing_value_table)
```

## 3c. Dropping unnecessary columns

```python
# creating a new dataframe and dropping the columns with more than 30% missing value
threshold = 0.3
missing_value_table['Percent Missing'] = missing_value_table['Percent Missing'].str.strip('%').astype(float)
missing_value_columns = missing_value_table[missing_value_table['Percent Missing'] >= threshold*100].index
df_new = df.drop(missing_value_columns, axis=1)
```
